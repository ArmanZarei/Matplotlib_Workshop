```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_arr = [0, 4, 6, 10]
y_arr = [1, 6, 3, 5]
y_arr2 = [5, 0, 2, 3]
```

```python
plt.plot(x_arr, y_arr) # (x-axis, y-axis)
plt.show() # Optional in notebook
```

```python
plt.plot(
    x_arr, 
    y_arr, 
    linestyle='--',
    linewidth=3,
    marker='o', 
    color='#3498db', # or c='#3498db'
    markerfacecolor='cyan', # or mfc='cyan'
    markeredgewidth=2, # or mew=2
    markeredgecolor='blue', # or mec='blue'
    markersize=10 # or ms=10
);
```

```python
plt.plot(x_arr, y_arr)
plt.plot(x_arr, y_arr2);
```

# Subplot

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1) # (row, col, idx)
plt.plot(x_arr, y_arr)
plt.title('First')
plt.subplot(1, 2, 2)
plt.plot(x_arr, y_arr2)
plt.title('Second')
plt.suptitle("Super Title");
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].plot(x_arr, y_arr)
axs[0].set_title('First')
axs[1].plot(x_arr, y_arr2)
axs[1].set_title('Second');
fig.suptitle('Super Title');
```

```python
from matplotlib.gridspec import GridSpec

gs = GridSpec(2, 2)
ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[0, 1])
ax3 = plt.subplot(gs[1, :])
ax1.plot(x_arr, y_arr)
ax2.plot(x_arr, y_arr)
ax3.plot(x_arr, y_arr)
plt.show()
```
